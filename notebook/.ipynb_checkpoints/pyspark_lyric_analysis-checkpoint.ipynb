```python
from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession
import pandas as pd
```

```python
conf = SparkConf()
sc = SparkContext(conf=conf)
spark = SparkSession.builder.appName("lyric Clustering").getOrCreate()
```

```python
lyricDF = spark.read\
.format("com.databricks.spark.csv")\
.option("header", "true")\
.option("inferSchema", "true")\
.load("lyric_new.csv")
```

```python
lyricDF.count()
```

```python
lyricDF = lyricDF.na.drop()
```

```python
lyricDF.count()
```

```python
lyricDF.show()
```

```python
from pyspark.ml.feature import HashingTF, IDF, Tokenizer
```

```python
tokenizer = Tokenizer(inputCol="lyrics", outputCol="tokens")
tokenized_lyrics = tokenizer.transform(lyricDF)
```

```python
tokenized_lyrics.show()
```

```python
from pyspark.ml.feature import StopWordsRemover
```

```python
remover = StopWordsRemover(inputCol="tokens", outputCol="filterd_tokens")
tokenized_lyrics = remover.transform(tokenized_lyrics)
```

```python
tokenized_lyrics.show()
```

```python
hashingTF = HashingTF(inputCol="filterd_tokens", outputCol="tf", numFeatures=20)
tfLyric = hashingTF.transform(tokenized_lyrics)
```

```python
idf = IDF(inputCol="tf", outputCol="tfidf")
idfModel = idf.fit(tfLyric)
tfidfLyric = idfModel.transform(tfLyric)
```

```python
tfidfLyric.select("song", "tfidf").show(25)
```

```python
from pyspark.ml.feature import Normalizer
from pyspark.ml.linalg import Vectors

normalizer = Normalizer(inputCol="tfidf", outputCol="features")
normLyrics = normalizer.transform(tfidfLyric)

from pyspark.ml.clustering import KMeans

# Trains a KMeans model.
kmeans = KMeans().setK(6).setMaxIter(20)
km_model = kmeans.fit(normLyrics)

clustersTable = km_model.transform(normLyrics)
```

```python
clustersTable.groupBy("prediction").count().show()
```

```python
centers = km_model.clusterCenters()
print("Cluster Centers: ")
for center in centers:
    print(center)
```

```python
clustersTable.select("features").take(5)
```

```python
from pyspark.ml.feature import PCA as PCAml
from pyspark.mllib.linalg import Vectors
```

```python
pca = PCAml(k=3, inputCol="features", outputCol="pca")
model = pca.fit(clustersTable)
transformed = model.transform(clustersTable)
```

```python
pl = transformed.select("song", "genre", "artist", "pca", "prediction").cache().toPandas()
```

```python
# lyricDF = spark.read.parquet("/databricks-datasets/news20.binary/data-001/training")
lyricDF.cache().count()

lyricDF = lyricDF.na.drop()

from pyspark.ml.feature import HashingTF, IDF, Tokenizer, CountVectorizer, StopWordsRemover
tokenizer = Tokenizer(inputCol="lyrics", outputCol="tokens")
remover = StopWordsRemover(inputCol="tokens", outputCol="stopWordsRemovedTokens")
hashingTF = HashingTF(inputCol="stopWordsRemovedTokens", outputCol="rawFeatures", numFeatures=2000)
idf = IDF(inputCol="rawFeatures", outputCol="features", minDocFreq=5)

from pyspark.ml.clustering import KMeans
kmeans = KMeans(k=20)

from pyspark.ml import Pipeline
pipeline = Pipeline(stages=[tokenizer, remover, hashingTF, idf, kmeans])

model = pipeline.fit(lyricDF)

results = model.transform(lyricDF)
results.cache()

# display(results.groupBy("prediction").count())  # Note "display" is for Databricks; use show() for OSS Apache Spark
```
